```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = np.arange(1, 10**7, 1)
```

```python
def Z_LC(R, f, L ,C):
    w = 2*np.pi*f
    return np.sqrt(R**2 + (w*L - 1/(w * C) )**2)
```

```python
def Z_C(R, f ,C):
    w = 2*np.pi*f
    return np.sqrt(R**2 + ( 1/(w * C))**2)
```

```python
def Z_L(R, f ,L):
    w = 2*np.pi*f
    return np.sqrt(R**2 + (w * L)**2)
```

```python
def theta(R, f, L ,C):
    w = 2*np.pi*f
    return np.arctan(-((w*L) - 1/(w*C) )/R)
```

```python
R = 1000
C = 10e-9
L = 1.85e-3
E = 10
```

```python
ER_LC =( R * E) / Z_LC(R, f, L ,C)
ER_L =( R * E) / Z_L(R, f, L)
ER_C =( R * E) / Z_C(R, f, C)
```

```python
plt.plot(f, ER_LC, c = 'b', label = 'Theoretical value')
plt.xlabel('Freqency[Hz]')
plt.ylabel('Volt [V]')
plt.title('RLC circuit')
# plt.scatter(f2, V, c= 'red')
plt.xscale('log')
plt.xlim(1,10**6)
plt.grid(True, which = "both", ls="--",color = "lightblue")
plt.legend()
plt.savefig('data1_1.png')
ER_LC.max()
```

```python
def TransientPhenomenon(t, E, R, C,t0 = 0):
    x = -t/(R*C)



    return E*np.exp(-(t-t0)/(R*C))
```

```python
Ep = 10.
Em = -10.
R = 1000.
C = 10e-10
t1 = np.linspace(0, 1/100000 * 1/2 , 11)
t2 = np.linspace(1/100000 * 1/2, 1/100000 , 11)

print(np.array(t1).shape)
```

```python
ERp = TransientPhenomenon(t1, Ep, R, C)
ERm = TransientPhenomenon(t2, Em, R, C, t0 = 1/100000 * 1/2)
```

```python
plt.plot(t1,ERp, label = 'Theoretical value', c = 'b')
plt.plot(t2,ERm, c = 'b')
plt.yticks(np.arange(-10,10,1))
plt.grid(True, c = 'lightblue')
plt.ylim(-10.1, 10.1)
plt.title('transient')
plt.xlabel('time[s]')
plt.ylabel('Volt [V]')
plt.legend()
plt.savefig('transient1.png')
```
